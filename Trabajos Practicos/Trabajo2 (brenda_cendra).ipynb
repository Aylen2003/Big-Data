### Ejercicio 1

* Â¿QuÃ© porcentaje de varones y de mujeres sobreviviÃ³ dentro de cada clase (primera, segunda, tercera)?

```python
# Respuesta acÃ¡.

# Calulo el  conteo total de pasajeros por gÃ©nero y clase

total_pasajeros = data_grouped.size()

# Calculo el conteo pasajeros que sobrevivieron por gÃ©nero y clase

sobrevivientes = data_grouped['survived'].sum()

# calculo el porcentaje de pasajeros que sobrevivieron por gÃ©nero y clase

procentaje_sobrevivientes = (sobrevivientes / total_pasajeros) * 100

# Inprimo los porcentaje de supervivencia

display(procentaje_sobrevivientes)
```

### Ejercicio 2

* Crear bines para las edades: (0-18] (18-80]

* Â¿QuÃ© porcentaje de pasajeros sobreviviÃ³, dado su sexo, franja etaria y clase?

```python
# Respuesta acÃ¡
# Creo los bines para las edades

bins = [0, 18, 80]
labels = ['0-18', '18-80']
data['age_group'] = pd.cut(data['age'], bins, labels=labels, right=True)

# Calculo el porcentaje de pasajeros que sobrevivieron segÃºn su sexo, franja etaria y clase
survived_rates = data.groupby(['sex', 'age_group', 'class'])['survived'].mean().reset_index()

# Renombro la columna para mayor claridad
survived_rates['survivedl'] = survived_rates['survived']  * 100

print(survived_rates)
```
