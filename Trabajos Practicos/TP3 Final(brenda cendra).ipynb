### Ejercicio 3

Obtener las empresas que no tienen registrado ningun empleado en el `DataFrame` `employee_df`

**Tips de funciones a utilizar:**
  - merge
  - group_by

````python
merged_df = company_df.merge(employee_df,how='left',left_on='company_id', right_on='company')

employee_count = merged_df.groupby('company_id')['employee_id'].count().reset_index()

companies_with_no_employees = employee_count[employee_count['employee_id'] == 0]

result = company_df[company_df['company_id'].isin(companies_with_no_employees['company_id'])]

print('Empresas sin empleados registrados:')
display(result)
````

### Ejercicio 5

Dados los `DataFrames` `df1` y `df2`, crear una columna que indique el tipo de join que se realizo para cada registro.
**Tips de funciones a utilizar:**
  - merge

````python
df1 = pd.DataFrame({"col1": [0, 1], "col_left": ["a", "b"]})
df2 = pd.DataFrame({"col1": [1, 2, 2], "col_right": [2, 2, 2]})

display(df1)
display(df2)


### Completa con tu c칩digo aqui

df_merged = pd.merge(df1, df2, on='col1', how='outer', indicator=True)

df_merged['_merge_type'] = df_merged['_merge'].map({
    'left_only': 'left_only',  
    'right_only': 'right_only', 
    'both': 'both'              
})

display(df_merged)

````

### Ejercicio 6

A partir de los `DataFrames` `left`,`right` y `right2`, obtener el resultado del `join` de las 3 tablas utilizando el campo `k` (para `left` y `right`) y el `index` (para `right2`) como se ve en la imagen

<img src="img/merging_join_multi_df.png" width="500">


**Tips de funciones a utilizar:**
  - set_index
  - join

````python
left = pd.DataFrame({"k": ["K0", "K1", "K2"], "v": [1, 2, 3]})
right = pd.DataFrame({"k": ["K0", "K0", "K3"], "v": [4, 5, 6]})
right2 = pd.DataFrame({"v": [7, 8, 9]}, index=["K1", "K1", "K2"])

### Completa con tu c칩digo aqui

result = left.set_index('k').join(right.set_index('k'),lsuffix='_x',rsuffix='_y')

result = result.join(right2,rsuffix='_z')

display(result)
````
